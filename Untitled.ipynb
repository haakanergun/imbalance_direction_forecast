```python
url = "https://seffaflik.epias.com.tr/transparency/service/market/smp?endDate=2023-12-31&startDate=2023-01-01"
response = requests.get(url)
data = response.json()
```

```python
base_url = "https://seffaflik.epias.com.tr/transparency/service/market/bpm-order-summary"
start_date = "2022-12-31"
end_date = "2023-12-31"

params = {
    "startDate": start_date,
    "endDate": end_date
}

dgp_response = requests.get(base_url, params=params)
dgp_data = dgp_response.json()
```

```python
# 'bpmOrderSummaryList' anahtarını kullanarak ilgili bilgilerin elde edilmesi
dgp_list = dgp_data['body']["bpmOrderSummaryList"]

# Tüm saatler için ilgili verileri içeren bir DataFrame oluşturulması
hour_data = []

for hour_info in dgp_list:
    hour_info["tarih"] = pd.to_datetime(hour_info["date"], format="%Y-%m-%dT%H:%M:%S.%f%z")
    hour_data.append(hour_info)

# Tüm saatlerin verilerini içeren tek bir DataFrame oluşturulması
dgp_df = pd.DataFrame(hour_data)
```
