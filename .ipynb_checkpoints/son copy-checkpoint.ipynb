# 2. Sezgisel Kuraln Performansnn Deerlendirilmesi

# 2.1. Sezgisel Kural Uygulanmas (T-3)

T-3 sezgisel kural, elektrik piyasalarnda kullanlan bir kuraldr ve dengesizlik tahminlerinde kullanlr. Elektrik piyasalar, arz ve talep dengesini srekli olarak salamak zorundadr, nk elektrik enerjisi annda tketilmesi gereken ve kolayca depolanamayan bir rndr. Dengesizlik, tketim ve retim arasndaki farkn olmas durumudur ve bu farklar, enerji sistemini istikrarsz hale getirebilir.

T-3 sezgisel kural, u anki saat (T) ile 3 saat nceki saat (T-3) arasndaki dengesizlik ynnn (arz fazlal veya talep fazlal) benzer olduunu varsayar. Bu kural, zellikle gn ii elektrik piyasasnda, dengesizlik ynetimi ve dengesizlik fiyatlarnn tahmin edilmesinde kullanlr. Mant, gemi saatlerdeki dengesizlik ynnn, mevcut saatteki dengesizlik yn zerinde belirleyici bir etkisi olduunu dnmektir.

Aada, belirtilen kuraln performans llm ve yorumlanmtr.

```python
df["predicted_smp_direction"] = df["smpDirection"].shift(3)
df = df.dropna() # Lag alndktan sonra ilk  veri hesaplanamayaca iin ve ayn zamanda metrik sonucunu etkileyecei iin drlmtr.
```

## 2.2. Performans Metriklerinin Hesaplanmas

Sezgisel kuraln performansnn hesaplanmas iin doruluk (accuracy) metrii kullanlmtr.

```python
accuracy = (df["smpDirection"] == df["predicted_smp_direction"]).mean()
print(f"Sezgisel kuraln performans: {accuracy:.2%}")
```

### 2.3. Saatlik ve Aylk Performans Analizi

Saatlik ve aylk performans analiz etmek iin, saat ve ay stunlar oluturulmu ve doruluklar hesaplanmtr.

```python
df["hour"] = df["date"].dt.hour
df["month"] = df["date"].dt.month

hourly_accuracy = df.groupby("hour").apply(lambda x: (x["smpDirection"] == x["predicted_smp_direction"]).mean())
monthly_hourly_accuracy = df.groupby(["month", "hour"]).apply(lambda x: (x["smpDirection"] == x["predicted_smp_direction"]).mean())

print("Saatlik performans:")
print(hourly_accuracy)

print("\nAylk bazda saatlik performans:")
print(monthly_hourly_accuracy.unstack())
```

## 3.1. EPA'tan "DGP Talimat REST Servisi" Verilerini ekme

EPA'tan "DGP Talimat REST SERVS" verilerini ekmek iin, gerekli parametreler ve request URL'si belirlenmi.

#### Parametreler

- Bu problem iin start_date 2018-01-01, end_date 2022-12-31 olarak belirlenmitir.
- Model, sezgisel kural ile kyaslanaca iin 2022 yl test veri kmesi olacaktr.
- 2021 validasyon veri kmesi olacak ve 2021den nceki veri eitim veri seti olacaktr.
- %60 eitim, %20 validasyon ve %20 test.

```python
base_url = "https://seffaflik.epias.com.tr/transparency/service/market/bpm-order-summary"
start_date = "2018-01-01"
end_date = "2022-12-31"

params = {
    "startDate": start_date,
    "endDate": end_date
}

dgp_response = requests.get(base_url, params=params)
dgp_data = dgp_response.json()
```

## 3.2. Veriyi Pandas DataFrame'e Dntrme

eitli n ilemelerin yaplmas iin veri pandas DataFrame'e dntrlmtr

```python
# 'bpmOrderSummaryList' anahtarn kullanarak ilgili bilgilerin elde edilmesi
dgp_list = dgp_data['body']["bpmOrderSummaryList"]

# Tm saatler iin ilgili verileri ieren bir DataFrame oluturulmas
hour_data = []

for hour_info in dgp_list:
    hour_info["tarih"] = pd.to_datetime(hour_info["date"], format="%Y-%m-%dT%H:%M:%S.%f%z")
    hour_data.append(hour_info)

# Tm saatlerin verilerini ieren tek bir DataFrame oluturulmas
dgp_df = pd.DataFrame(hour_data)
```

## 3.3. zellik karm

### 3.3.1. Covid Flag

TimeSeries problemlerinde Covid-19 dnemine Flag atamak doru bir yaklam olabilir.

```python
dgp_df["covid_flag"] = np.where((dgp_df["date"] >= "2020-03-11") & (dgp_df["date"] <= "2022-05-31"), 1, 0)
```

### 3.3.2 Tarihsel zellik karm

```python
dgp_df['Year'] = dgp_df['tarih'].dt.year
dgp_df['Month'] = dgp_df['tarih'].dt.strftime('%B')
dgp_df['Day'] = dgp_df['tarih'].dt.day
dgp_df['Day of Week'] = dgp_df['tarih'].dt.strftime('%A')
dgp_df['Day of Year'] = dgp_df['tarih'].dt.day_of_year
dgp_df['Series'] = np.arange(0,len(dgp_df))+1
dgp_df['decmonth'] = dgp_df['tarih'].dt.month
dgp_df["hour"] = dgp_df["tarih"].dt.hour
# Gn balanglarnn belirlenmesi
dgp_df['day_start'] = dgp_df['hour'] == 0
```

### 3.3.3 net Deikeninden zellik karmlar

net deikeninin gnlk bazda kmlatif ortalamas ve bir saatten 180 saate (yaklak bir hafta) lag'nn alnmas doru bir yaklam olabilir.

```python
# Kmlatif ortalama stununun tanmlanmas
dgp_df['cumulative_avg'] = 0.0

# Gnlerin balanglarnn alnmas
day_starts = dgp_df[dgp_df['day_start']].index.tolist()

# Her gnn kmlatif ortalamasn hesaplama
for i in range(len(day_starts)):
    start = day_starts[i]
    if i < len(day_starts) - 1:
        end = day_starts[i + 1]
    else:
        end = len(dgp_df)
    cum_sum = 0
    count = 0
    for j in range(start, end):
        cum_sum += dgp_df.loc[j, 'net']
        count += 1
        dgp_df.loc[j, 'cumulative_avg'] = cum_sum / count

for lag in range(1, 180):
    dgp_df[f"net_lag{lag}"] = dgp_df["net"].shift(lag)
```

## 3.4. D Verilerin Eklenmesi

### 3.4.1 zel Gnler

zel gnler, zaamn serisi analizinde olduka faydaldr. Belirli bir tarihsel veri erevesinde zel gnlerin varl, bu zel gnlerin sat, trafiin younluu, vb. gibi farkl verilerin analizinde kullanlabilir.

```python
df = dgp_df.copy()
dfs = pd.read_html('http://www.bulanik.gov.tr/ulkemizde-kutlanan-onemli-gun-ve-haftalar')
i_days = pd.concat(dfs)
conv_months = {'Ocak':1,'ubat':2,'Mart':3,'Nisan':4,'Mays':5,'Haziran':6,'Temmuz':7,'Austos':8,'Eyll':9,'Ekim':10,'Kasm':11,'Aralk':12}
i_days = i_days[i_days[0].str.contains('Gn')]
i_days=i_days[~i_days[1].str.contains('ilk')]
i_days = i_days[~i_days[1].str.contains('-')]
new_months = i_days[1].str.split(' ',expand=True)[1].replace(conv_months)

i_days['days'] = i_days[1].str.split(' ',expand=True)[0]
i_days['months'] = new_months

i_days['days']=i_days['days'].astype('int64')

df['Year'] = df['tarih'].dt.year
df['Month'] = df['tarih'].dt.strftime('%B')
df['Day'] = df['tarih'].dt.day
df['Day of Week'] = df['tarih'].dt.strftime('%A')
df['Day of Year'] = df['tarih'].dt.day_of_year
df['Series'] = np.arange(0,len(df))+1
df['decmonth'] = df['tarih'].dt.month
```

```python
new_df = pd.merge(df, i_days,  how='left', left_on=['decmonth','Day'], right_on = ['months','days'])
new_df = new_df.drop([1,'months'],axis=1).rename(columns={0:'Important Days'})
new_df['Important Days'] = new_df['Important Days'].fillna('-')
```

Buradaki ama, daha sonra toplanacak havadurumu veri seti iin hangi iller baznda hava durumu, ya ve rzgar verisinin toplanacana karar vermektir. Btn illerin verisini almak, overfitting problemine yol aabilecei iin yalnzca bu ehirler toplanmtr. ncelikle bu listeler enerjiatlasi.com sitesinden alnmtr ve daha sonra iller listelenerek bu iller baznda havadurumu,rzgar,ya ve gne ile alakal veriler ekilmitir

```python
def scrape_table_from_url(url, type):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table_html = str(soup.find_all('table'))
    if type == 'gunes':
        df = pd.read_html(table_html)[0]
    elif type == 'ruzgar':
        df = pd.read_html(table_html)[0]
    elif type == 'hidroelektrik':
        df = pd.read_html(table_html)[0]
    return df


url = 'https://www.enerjiatlasi.com/en-buyuk/gunes'
df_gunes = scrape_table_from_url(url, 'gunes')

url = 'https://www.enerjiatlasi.com/en-buyuk/ruzgar'
df_ruzgar = scrape_table_from_url(url, 'ruzgar')

url = 'https://www.enerjiatlasi.com/en-buyuk/hidroelektrik'
df_hidroelektrik = scrape_table_from_url(url, 'hidroelektrik')

df = pd.concat([df_ruzgar, df_hidroelektrik], ignore_index=True)
```

## 4.1. zellik Seiminin Yaplmas

(Bu ksm da uzun srd iin, pickle ile daha nceden kaydedilmi boruta modelini import edebilirsiniz)

Boruta, Rastgele Orman Snflandrc algoritmas temelinde alan bir zellik seimi yntemidir. Boruta algoritmasnn arkasndaki fikir hem ilgin hem de basittir: orijinal veri kmesindeki tm zelliklerin rastgele kopyalarn olutururuz (glge zellikler olarak adlandrlr) ve bu geniletilmi veri kmesine dayal olarak snflandrclar eitiriz. Bir zelliin nemini anlamak iin, onu oluturduumuz tm glge zelliklerle karlatrrz. Sadece bu sentetik zelliklerden istatistiksel olarak daha nemli olan zellikleri tutarz, nk bunlar model performansna daha fazla katkda bulunurlar.

Boruta algoritmas u admlar izler:

- Orijinal veri kmesindeki tm zellikleri kopyalar ve bunlar orijinal zelliklerle birletirir.
- Glge zelliklere rastgele deerler atar ve bunlar orijinal zelliklerden ayrt etmek iin adlarn deitirir.
- Rastgele Orman Snflandrc algoritmasn geniletilmi veri kmesine uygular ve her bir zelliin nemini hesaplar.
- Glge zelliklerin maksimum nemini alr ve bunu bir eik deeri olarak kullanr.

```python
from boruta import BorutaPy
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier


var = 'target_direction'

X = merged_df.drop([var,'date','nextHour','tarih'], axis=1)
y = merged_df[var]

rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', n_estimators=200, max_depth=7, min_samples_split=5)
boruta_feature_selector = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=1, max_iter=100, perc=100)

categorical_columns = X.select_dtypes('object').columns.tolist()  # Kategorik stun adlarn liste halinde salayn

label_encoder = LabelEncoder()

for column in categorical_columns:
    X[column] = label_encoder.fit_transform(X[column])

# Boruta ile zellik seimini gerekletirin
boruta_feature_selector.fit(np.array(X), np.array(y))

import pickle

# Boruta seicisini 'importances' klasrne kaydedin
with open(f'{var}_boruta_selector2.pkl', 'wb') as f:
    pickle.dump(boruta_feature_selector, f)

print(f"{var} iin Boruta seici kaydedildi.")
```

```python
def yuzdeler(model, df, prediction_label, y):
    preds = predict_model(model, df)[[y, 'prediction_label', 'prediction_score']]
    son = pd.DataFrame()
    preds = preds[preds['prediction_label']==prediction_label]
    arange = np.arange(0.3, 1, 0.005)
    for i in arange:
        deneme = preds[preds['prediction_score']>=i]
        yuzde = round(accuracy_score(deneme[y], deneme['prediction_label']), 3)
        dicto = {'Thresh': i, 'Yzde': yuzde, 'Len': len(deneme)}
        son = son.append(pd.DataFrame(dicto, index=[0]), ignore_index=True)
        son = son.dropna(axis=0)
    #son.to_csv(f"precs/{y}model_{prediction_label}.csv", index=False)
    return son.sort_values('Thresh', ascending=False)
```

Bu fonksiyon, bir snflandrma modelinin performansn farkl eik (threshold) deerleri iin yzde doruluk (accuracy) oran olarak deerlendiren bir fonksiyondur. Fonksiyon, model, veri erevesi (df), tahmin edilen etiket (prediction_label) ve gerek etiket (y) gibi argmanlar alr.

Modelin rettii threshold ise softmax fonksiyonu ile retilir. Softmax fonksiyonu, modelin ktlarn olaslk dalmna dntrmek iin kullanlr. Enerji piyasalarnn gelecekteki durumu iin birden fazla snf tahmin etmek istediimizde, softmax fonksiyonu her snf iin bir olaslk deeri atar. Bu olaslk deerleri, tahmin edilen snflarn gven dzeyini gsterir ve tm snflarn olaslklar toplam 1'e eittir.
